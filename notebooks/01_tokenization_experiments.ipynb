# Tokenization Experiments

## Goal
Different tokenizer make different decisions.
This notebook explores how tokenizer choice affects the output
and where each approach breaks.

```python
import re
from collections import Counter
```

## 1. Word Tokenization
The simplest approach - split by whitespace.

```python
def whitespace_tokenizer(text: str) -> list:
    """Split text by whitespace - the most naive approach."""
    return text.split()

# Test cases
tests = [
    "I love NLP",
    "Don't you think it's great?",
    "U.S.A is a country",
    "#MachineLearning is trending",
    "This costs $9.99 today"
]

for text in tests:
    tokens = whitespace_tokenizer(text)
    print(f"Input: {text}")
    print(f"Tokens: {tokens}")
    print()
```

## 2. Regex Tokenizer
More contorl - define exactly what a token looks like.

```python
def regex_tokenizer(text: str) -> list:
    """Extract only alphabetic words using regex."""
    return re.findall(r"\b[a-zA-Z]+\b", text)

for text in tests:
    tokens = regex_tokenizer(text)
    print(f"Input: {text}")
    print(f"Tokens: {tokens}")
    print()
```

## 3. Observations

| Tokenizer | Strenght | Waekness |
|-----------|----------|----------|
| Whitespace | Simple, fast | Punctuation sticks to words |
| Regex | Clean output | Contractions and abbreviations break |

## Key Insight
No simple tokenizer handles all edge cases.
Production systems use trained tokenizers (spaCy, HuggingFace)
that learned from real text how to handle these cases.

## What breaks every simple tokenizer
- Contractions: "don't", "It's", "I'm"
- Accreviations: "U.S.A", "Dr.", "etc."
- Special tokens: hashtage, emails, "URLs, prices
- Languages without spaces: Chinese, Japanese
