# Stemming vs Lemmatization

## Goal 
Both reduce words to their root form.
This notebook explores the difference in practice
and shows when each approach matters.

```python
import re
```

## 1. Rule-Based Stemming
Chops off suffixes using fixed rules.
Fast but rough - output may not be a real word.

```python
def simple_stemmer(word: str) -> str:
    """Rule-based stemmer - strips common suffixes."""
    rules = [
        (r"ing$", ""),
        (r'tion$', ''),
        (r'ed$', ''),
        (r'ly$', ''),
        (r'ies$', 'y'),
        (r'ness$', ''),
        (r'er$', ''),
        (r'est$', ''),
        (r'ment$', ''),
        (r's$', ''),
    ]
    for pattern, replacement in rules:
        if re.search(pattern, word):
            return re.sub(pattern, replacement, word)
    return word

# Test
test_words = [
    "running", "studies", "happiness",
    "quickly", "eating", "studies",
    "better", "greatest", "caring"
]

print(f"{"word"}:<15 {"Stemmed":<15}")
print("-" * 30)
for word in test_words:
    print(f"{word:<15} {simple_stemmer(word):<15}")
```

## 2. Rule-Based Lemmatization
Uses a lookup table instead of just chopping suffixes.
Slower but produces real words.

```python
# Simple lookup-based lemmatizer
LEMMA_LOOKUP = {
    'running': 'run', 'runs': 'run', 'ran': 'run',
    'studies': 'study', 'studied': 'study', 'studying': 'study',
    'happiness': 'happy', 'happily': 'happy',
    'quickly': 'quick', 'quicker': 'quick', 'quickest': 'quick',
    'eating': 'eat', 'ate': 'eat', 'eaten': 'eat',
    'better': 'good', 'best': 'good',
    'greatest': 'great', 'greater': 'great',
    'caring': 'care', 'cared': 'care', 'cares': 'care',
}

def lookup_lemmatizer(word: str) -> str:
    """Lookup-based lemmatizer - returns base from dictionary."""
    return LEMMA_LOOKUP.get(word, word)

print(f"{"word":<15} {"Stemmed":<15} {"Lemmatized":<15}")
print("-" * 45)
for word in test_words:
    print(f"{word:<15} {simple_stemmer(word):<15} {lookup_lemmatizer(word):<15}")
```

## 3. Observations

| Word | runn | run |
| happiness | happi | happy |
| better | bett | good |
| caring | car | care |

## Key Insight
Stemming is fast but produces non-words.
Lemmatization produces real words but needs a lookup table or trained model.

For production NLP:
- Search engines -> stemming is usually enough
- Text understanding, sentiment, classification -> lemmatization is better

## Limitation of this experiment
Our lookup-besed lemmatizer only knows words we defined.
Real lemmatizers (spaCy, WordNet) know hundreds of thousands of words.
